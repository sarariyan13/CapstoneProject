```python
# Write a query to retrieve all shoe names and their corresponding prices for men's shoes

query = ''' 
SELECT Shoes_names, shoes_price
FROM Data1
WHERE shoes_category = 'Men'; 
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
# Write a query to retrieve the number of different colors available for each category
query = '''
SELECT
    shoes_category,
    COUNT(DISTINCT shoes_color) AS num_colors
FROM
    Data1
GROUP BY
    shoes_category;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```

```python
# Write a query to find the cheapest women's shoe in a specific color (e.g., 'Black ').

query = '''
SELECT Shoes_names, shoes_price
FROM Data1
WHERE shoes_Category = 'Women' AND (
    LOWER(shoes_color) LIKE '%black%' OR
    LOWER(shoes_color) LIKE '%/black%' OR
    LOWER(shoes_color) LIKE '%black/%'
)
;
'''
result_df = pd.read_sql_query(query, conn)
result_df
```
